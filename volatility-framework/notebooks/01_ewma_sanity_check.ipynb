```python
import numpy as np
from volaframe.models import EWMAVolatilityModel

returns = np.random.normal(0, 0.01, size=1000)

model = EWMAVolatilityModel(lambda_=0.94)
model.fit(returns)

forecast = model.predict(h=5)
forecast
```

```python
from volaframe.metrics import qlike

# fake realized variance
rv = returns**2

# EWMA forecasted variance
sigma_hat = model.volatility_**2

ql = qlike(rv[-len(sigma_hat):], sigma_hat)
ql.mean()
```
