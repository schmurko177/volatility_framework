```python
import numpy as np

from volaframe.models import EWMAVolatilityModel, RollingVarianceModel
from volaframe.metrics import qlike
```

```python
np.random.seed(42)
returns = np.random.normal(0, 0.01, size=2000)

rv = returns**2
```

```python
ewma = EWMAVolatilityModel(lambda_=0.94)
ewma.fit(returns)

ewma_var = ewma.volatility_**2
```

```python
rolling = RollingVarianceModel(window=21)
rolling.fit(returns)

rolling_var = rolling.volatility_**2
```

```python
# align lengths
min_len = min(len(ewma_var), len(rolling_var))

ql_ewma = qlike(rv[-min_len:], ewma_var[-min_len:])
ql_roll = qlike(rv[-min_len:], rolling_var[-min_len:])

ql_ewma.mean(), ql_roll.mean()
```
