```python
import numpy as np
from volaframe.models import EWMAVolatilityModel
from volaframe.metrics import qlike
```

```python
# fake returns
returns = np.random.normal(0, 0.01, size=1500)

# EWMA model
model = EWMAVolatilityModel(lambda_=0.94)
model.fit(returns)

# 1-step forecast
sigma_hat = model.volatility_**2
rv = returns**2

ql = qlike(rv[-len(sigma_hat):], sigma_hat)

ql.mean()
```
